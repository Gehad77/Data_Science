```python
# modes_examples.py
# "r" قراءة
with open("te.txt", "r", encoding="utf-8") as f:
    print("r:", f.readline().strip())

# "w" كتابة (يمسح المحتوى لو موجود)
with open("tmp_w.txt", "w", encoding="utf-8") as f:
    f.write("كتابة جديدة\n")

# "a" إضافة
with open("tmp_a.txt", "a", encoding="utf-8") as f:
    f.write("سطر مضاف\n")

# "x" إنشاء جديد وفشل لو موجود
try:
    with open("tmp_x.txt", "x", encoding="utf-8") as f:
        f.write("ملف جديد\n")
except FileExistsError:
    print("tmp_x.txt موجود بالفعل")

# ثنائي "rb" و "wb"
with open("tmp_bin.bin", "wb") as f:
    f.write(b"\x00\x01\x02")

# "r+" قراءة وكتابة
with open("tmp_rw.txt", "w+", encoding="utf-8") as f:
    f.write("abc\n")
    f.seek(0)
    print("r+ read:", f.read())
```

```python
#قراءة كل المحتوى: read()
# read_full.py
with open("te.txt", "r", encoding="utf-8") as f:
    content = f.read()
print(content)
```

```python
#قراءة سطر-سطر: readline()
# read_line_by_line_readline.py
with open("te.txt", "r", encoding="utf-8") as f:
    while True:
        line = f.readline()
        if not line:
            break
        print(line.strip())
```

```python
#قراءة كل الأسطر في قائمة: readlines()
# readlines_example.py
with open("te.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
for i, line in enumerate(lines):
    print(i, line.strip())
```

```python
#التكرار المباشر على كائن الملف (الأكفأ للسطر-سطر)

# iterate_file.py
with open("te.txt", "r", encoding="utf-8") as f:
    for line in f:
        print(line.strip())
```

```python
#كتابة نص (write) وكتابة قائمة (writelines)

# write_examples.py
with open("test1.txt", "w", encoding="utf-8") as f:
    f.write("Hello\n")

lines = ["سطر1\n", "سطر2\n"]
with open("test1_lines.txt", "w", encoding="utf-8") as f:
    f.writelines(lines)
```

```python
# الإضافة (append) مع حلقة لإضافة 30 سطر

# append_30.py
with open("te.txt", "a", encoding="utf-8") as f:
    for i in range(30):
        f.write(f"I love python {i}\n")
```

```python
# tell() و seek()
# tell_seek_example.py
with open("te.txt", "r", encoding="utf-8") as f:
    print("موقع بداية:", f.tell())
    part = f.read(20)
    print("بعد قراءة 20 حرف:", f.tell())
    f.seek(5)  # اذهب للموضع 5
    print("بعد seek لموضع 5:", f.tell())
    print("من هناك:", f.read(10))
```

```python
# قراءة الملفات الكبيرة بقطع (chunks) لتجنب استهلاك الذاكرة

# read_in_chunks.py
def read_in_chunks(path, chunk_size=1024):
    with open(path, "r", encoding="utf-8") as f:
        while True:
            data = f.read(chunk_size)
            if not data:
                break
            yield data

for chunk in read_in_chunks("large.txt", 4096):
    process = chunk  # بدلها معالجة فعلية
    print("chunk length:", len(process))
```

```python
# التعامل مع الأخطاء (FileNotFoundError, IOError)

# safe_read.py
try:
    with open("noexist.txt", "r", encoding="utf-8") as f:
        print(f.read())
except FileNotFoundError:
    print("الملف غير موجود")
except IOError as e:
    print("خطأ في الإدخال/الإخراج:", e)
```
